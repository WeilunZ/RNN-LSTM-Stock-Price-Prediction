```python
import pandas as pd
import numpy as np
import keras  # i Importing required keras packages
from keras.layers import Dense
from keras.models import Sequential
import matplotlib.pyplot as plt
import math
```

```python
class Model:
    def __init__(self,input_dim,optimizer_function,loss_function):
        self.input_dim=input_dim
        self.optimizer_function=optimizer_function
        self.loss_function=loss_function
    def getModel(self):
        model=Sequential()
        # We have 2 hidden layers
        model.add(Dense(12, activation='relu', input_dim=self.input_dim))  # Setting up 1st Dense layers where each node is connected to every node in the next layer
#         model.add(Dense(32, activation='relu'))  # Setting up 2nd layer
        # 1 output layer
        model.add(Dense(1))  # Setting up final layer
        
        model.compile(optimizer=self.optimizer_function, loss=self.loss_function)
        return model
```

```python
def train_test_split(X_data,y_data,train_percentage):
    n_train_intervals = math.ceil(X_data.shape[0] * train_percentage) 
    X_train,y_train=X_data[:n_train_intervals],y_data[:n_train_intervals]
    X_test,y_test=X_data[n_train_intervals:],y_data[n_train_intervals:]
    return X_train,y_train,X_test,y_test

def data_to_x_y(data,n_in_steps,n_out_steps):
    X_data,y_data=[],[]
    for i in range(0,data.shape[0]-n_in_steps-n_out_steps):
        X_data.append(data[i:i+n_in_steps])
        y_data.append(data[i+n_in_steps+n_out_steps-1])
    X_data=np.array(X_data)
    y_data=np.array(y_data)
#     X_data=X_data.reshape(X_data.shape[0],X_data.shape[1],1)
#     y_data=y_data.reshape(y_data.shape[0],1)
    return X_data,y_data
```

```python
stock_name='GOOGL'
```

```python
filepath='../../dataset/5year/'+stock_name+'-5year.csv'
df_ori=pd.read_csv(open(filepath),header=0)
df_ori.index = pd.to_datetime(df_ori['Date'])  # 将字符串索引转换成时间索引
df_ori.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(0, 1))
sc_close=sc.fit_transform(df_ori['Close'].values.reshape(-1,1))
```

```python
timestep=30
future_day=15
```

```python
X_data,y_data=data_to_x_y(sc_close,timestep,future_day)
X_train,y_train,X_test,y_test=train_test_split(X_data,y_data,0.7)
X_train=X_train.reshape(X_train.shape[0],X_train.shape[1])
X_test=X_test.reshape(X_test.shape[0],X_test.shape[1])
print(X_train.shape)
print(y_train.shape)
```

```python
input_dim=timestep
optimizer_function='adam'
loss_function='mean_squared_error'
n_batch=20
n_epochs=50
```

```python
model=Model(input_dim,optimizer_function,loss_function).getModel()
history=model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=n_epochs,batch_size=n_batch,verbose=2,shuffle=False)
plt.plot(history.history["loss"] , label="Train Loss (%s)" % 'value')
plt.plot(history.history["val_loss"], label="Test Loss (%s)" % 'value')
plt.legend()
plt.savefig('../../image/BP-network/'+stock_name+str(future_day)+'step_trainloss.png')
plt.show()
```

```python
yhat=model.predict(X_test)
inv_yhat=sc.inverse_transform(yhat.reshape(-1,1))
inv_ytest=sc.inverse_transform(y_test.reshape(-1,1))
plt.figure(figsize=(15,10))
plt.plot(inv_ytest,label='Origin')
plt.plot(inv_yhat,label='Predict')
plt.legend(
    loc = 'upper center',
    bbox_to_anchor = (0.5, -0.05),
    fancybox = True,
    shadow = True,
    ncol = 5,
)
plt.title('BP-network prediction - %s'%stock_name)
plt.savefig('../../image/BP-network/'+stock_name+'result'+str(future_day)+'.png')
plt.show()
```

```python
#误差统计
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import math

avg=np.mean(inv_ytest)

rmse=math.sqrt(mean_squared_error(inv_yhat, inv_ytest))
mae=mean_absolute_error(inv_yhat, inv_ytest)                     
rmse_error_percentage = rmse/avg
mae_error_percentage = mae/avg
print('TEST RMSE: %.4f' % rmse)
print('TEST MAE: %.4f'% mae)
print("Test RMSE Error Percentage: %.2f/100.00" % (rmse_error_percentage * 100))
print("Test MAE Error Percentage: %.2f/100.00" % (mae_error_percentage * 100))
```
