```python
def data_to_x_y(data,n_in_steps,n_out_steps):
    X_data,y_data=[],[]
    for i in range(0,data.shape[0]-n_in_steps-n_out_steps):
        X_data.append(data[i:i+n_in_steps])
        y_data.append(data[i+n_in_steps+n_out_steps-1])
    X_data=np.array(X_data)
    y_data=np.array(y_data)
    X_data=X_data.reshape(X_data.shape[0],X_data.shape[1],1)
    y_data=y_data.reshape(y_data.shape[0],1)
    return X_data,y_data
```

```python
def train_test_split(X_data,y_data,train_percentage):
    n_train_intervals = math.ceil(X_data.shape[0] * train_percentage) 
    X_train,y_train=X_data[:n_train_intervals],y_data[:n_train_intervals]
    X_test,y_test=X_data[n_train_intervals:],y_data[n_train_intervals:]
    return X_train,y_train,X_test,y_test
```

```python
class LossHistory(Callback):
    def on_train_begin(self, logs={}):
        self.losses = []
    def on_batch_end(self, batch, logs={}):
        self.losses.append(logs.get('loss'))
```

```python
stock_name='HD'
```

```python
filepath='../../dataset/5year/'+stock_name+'-5year.csv'
df=pd.read_csv(open(filepath),header=0)
future_day=10
timestep=20
df.head()
```

```python
minmax = MinMaxScaler().fit(df.iloc[:, 4].values.reshape(-1,1))
df_log = minmax.transform(df.iloc[:, 4].values.reshape(-1,1))
X_data,y_data=data_to_x_y(df_log,timestep,future_day)
X_train,y_train,X_test,y_test=train_test_split(X_data,y_data,0.7)
```

```python
units=8
dropout_rate=0.4
input_shape=(timestep,1)
optimizer_function='adam'
n_batch=20
n_epochs=40
```

```python
print(X_train.shape)
model=Model(units,dropout_rate,input_shape).getModel()
model.compile(loss='mean_absolute_error', optimizer='adam')
his = LossHistory()
history = model.fit(X_train, y_train,validation_data=(X_test,y_test),epochs=n_epochs, 
                    batch_size=n_batch,verbose=2,callbacks=[his],shuffle=False)
plt.plot(history.history["loss"] , label="Train Loss (%s)" % 'value')
plt.plot(history.history["val_loss"], label="Test Loss (%s)" % 'value')
plt.legend()
# plt.savefig('../../image/EMD/'+stock_name+'/trainloss/'+str(future_day)+'step_trainloss%d.png'% i)
plt.show()
```

```python
# print(his.losses)
```

```python
inv_residue=minmax.inverse_transform(np.array(his.losses).reshape(-1,1))
resid_std=np.std(inv_residue)
print(resid_std)
```

```python
#test
ypre=model.predict(X_test,batch_size=n_batch,verbose=1)
inv_yhat=minmax.inverse_transform(ypre.reshape(-1,1))
inv_ytest=minmax.inverse_transform(y_test.reshape(-1,1))

upper_list,lower_list,rvalue_list,pvalue_list=[],[],[],[]
for i in range(inv_yhat.shape[0]):
    pvalue=inv_yhat[i]
    upper_bound=pvalue+1.96*resid_std*math.sqrt(future_day)
    lower_bound=pvalue-1.96*resid_std*math.sqrt(future_day)
    rvalue=inv_ytest[i]
    #print(i,upper_bound,lower_bound,rvalue)
    upper_list.append(upper_bound[0])
    lower_list.append(lower_bound[0])
    rvalue_list.append(rvalue[0])
    pvalue_list.append(pvalue[0]) 
    
plt.figure(figsize=(15,10))
plt.plot(inv_ytest,label='Origin')
plt.plot(inv_yhat,label='Predict')
plt.legend(
    loc = 'upper center',
    bbox_to_anchor = (0.5, -0.05),
    fancybox = True,
    shadow = True,
    ncol = 5,
)
plt.title(' Point Prediction by LSTM - %s'%stock_name)
plt.savefig('../../image/Interval/'+stock_name+str(future_day)+'point_prediction.png')
plt.show()
```

```python
plt.figure(figsize = (15, 10))
plt.fill_between([k for k in range(len(upper_list))],upper_list,lower_list,label='interval',alpha=0.5)
plt.plot(rvalue_list,label='real value',color='orange')
#plt.plot(pvalue_list,label='predict value',color='green')
plt.legend(
    loc = 'upper center',
    bbox_to_anchor = (0.5, -0.05),
    fancybox = True,
    shadow = True,
    ncol = 5,
)
plt.title('interval prediction stock market-%s'%stock_name)
plt.savefig('../../image/Interval/'+stock_name+'+10step.png')
plt.show()
```

```python
#PICP
cnt=0
sum_width=0
for i in range(len(rvalue_list)):
    rvalue=rvalue_list[i]
    upper_value=upper_list[i]
    lower_value=lower_list[i]
    width=upper_value-lower_value
    sum_width+=width
    if rvalue<=upper_value and rvalue>=lower_value:
        cnt+=1
picp=cnt/len(rvalue_list)
mpiw=sum_width/len(rvalue_list)
print("PICP:",picp)
#MPIW
print("MPIW:",mpiw)
#CWC
gamma=0 if picp>=0.95 else 1
cwc=mpiw*(1+gamma*picp*math.exp(-50*(picp-0.95)))
print("CWC:",cwc)
```
